```python
products=browser.find_elements_by_css_selector('li.product-base')
len(products)
```

```python
for item in products:
    print(item)
```

```python
df_product=pd.DataFrame(columns=['name','description','price','discount'])
for item in products:
    name=item.find_element_by_css_selector('div.product-brand')
    name=name.get_attribute('innerHTML')
    
    desc=item.find_element_by_css_selector('h4.product-product')
    desc=desc.get_attribute('innerHTML')
    
    discount=item.find_element_by_css_selector('span.product-discountPercentage')
    discount=discount.get_attribute('innerHTML')
    
    #when we cannot point to the inner structure consider its parent ,so in case of discounted price ,
    #take parent ,because discounted is not common 
    #price=discount=item.find_element_by_css_selector('div.product-price')
   # print(price.get_attribute('innerHTML'))

    price_parent=item.find_element_by_css_selector('div.product-price')
    price_child=price_parent.find_element_by_css_selector('span')
    try:
        price=price_child.find_element_by_css_selector('span.product-discountedPrice')
        price=price=price.get_attribute('innerHTML')
    except:
        price=price_child.get_attribute('innerHTML')
    curr_row={'name':name,
        'description':desc,
        'price':price,
        'discount':discount
     }
    df_product=df_product.append(curr_row,ignore_index=True)
df_product
    
    
   
```

```python
## by creating a function
def get_info(job,selector,attribute):
    
    try:
        
        tag=job.find_element_by_css_selector(selector)
    
        tag=tag.get_attribute(attribute)
    
    except:
        
        tag=None
        
    return tag

df_jobs=pd.DataFrame(columns=['title','company','salary','date','location','experience','keyskills','description','url'])

for job in jobs:
    
    job_title=get_info(job,'li.desig','title')
    
    company_name=get_info(job,'span.org','innerHTML')
    
    salary=get_info(job,'span.salary','innerHTML')
    
    date=get_info(job,'span.date','innerHTML')
    
    try:
        parent_location=job.find_element_by_css_selector('span.loc')
        location=get_info(parent_location,'span','innerHTML')
    except:
        location=None
    
    
    experience=get_info(job,'span.exp','innerHTML')
    
    
    keyskills=get_info(job,'span.skills','innerHTML')
    
    
    description=get_info(job,'span.desc','innerHTML')
    
    url=get_info(job,'a.content','href')
    
    curr_job={'title':job_title,
              'company':company_name,
              'salary':salary,
              'date':date,
              'location':location,
              'experience':experience,
              'keyskills':keyskills,
              'description':description,
               'url':url}
    
    df_jobs=df_jobs.append(curr_job,ignore_index=True)
    
df_jobs.head()
    
    
    
    
    

```

```python
browser=webdriver.Chrome("chromedriver.exe")
browser.get("https://www.goodreads.com/choiceawards/best-books-2017")
```

```python
books=browser.find_elements_by_css_selector('div.category.clearFix')
len(books)
```

```python
df_books=pd.DataFrame(columns=['category','book_name','url_to_book_image','url_to_book_description'])

for book in books:

    ##names of category
    category=book.find_element_by_css_selector('h4')
    category=category.get_attribute('innerHTML')
    

    #names of book
    book_name=book.find_element_by_css_selector('img.winner')
    book_name=book_name.get_attribute('alt')
    
    #url to book cover image
    url1=book.find_element_by_css_selector('img.winner')
    url1=url1.get_attribute('src')
    
    
    
    #url to book description
    url2=book.find_element_by_css_selector('a')
    url2=url2.get_attribute('href')
    
    #print(category,'|',book_name,' | ',url1,' | ',url2)
    
    curr_books={'category':category,
             'book_name':book_name,
             'url_to_book_image':url1,
             'url_to_book_description':url2
               }
    df_books=df_books.append(curr_books,ignore_index=True)
df_books.head()
    
    
```

```python
#2.a)	Author name
for url in df_books['url_to_book_description']:
    browser.get(url)
    full_author=browser.find_element_by_css_selector('span[itemprop=name]')
    
    full_author=full_author.get_attribute('innerHTML')
    print(full_author)
    print('----------------')
    
```

```python
#2.b)	Short description about the book
for url in df_books['url_to_book_description'].head(3):
    browser.get(url)
    print(browser)
    short_desc=browser.find_element_by_css_selector('div.readable.stacked.gcaBookDescription')
    short_desc=full_desc.text
    print(full_desc)
    print('----------------')
```

```python
#2.c)	Votes for the winner
for url in df_books['url_to_book_description']:
    browser.get(url)
    votes=browser.find_element_by_css_selector('span.greyText.gcaNumVotes')
    votes=votes.get_attribute('innerHTML')
    print(votes)
    print('------------------')
```

```python
#2 d)	Total votes for all the books under that category
for url in df_books['url_to_book_description']:
    browser.get(url)
    total_votes=browser.find_element_by_css_selector('greytext.gcaNumVotes')
    total_votes=total_votes.get_attribute('innerHTML')
    print(total_votes)
    print('-------------------------')
```

```python
#3.	Combine all the above mentioned information about each book and create a pandas data frame
df_books2=pd.DataFrame(columns=['description','author_name','votes_for_winner','Total_votes_for_all'])

for url in df_books['url_to_book_description']:
    browser.get(url)
    
    full_desc=browser.find_element_by_css_selector('div.readable.stacked.gcaBookDescription')
    full_desc=full_desc.text
    
    author_name=browser.find_element_by_css_selector('span[itemprop=name]')
    author_name=author_name.get_attribute('innerHTML')
    
    votes=browser.find_element_by_css_selector('span.greyText.gcaNumVotes')
    votes=votes.get_attribute('innerHTML')
    
    total_votes=browser.find_element_by_css_selector('greytext.gcaNumVotes')
    total_votes=total_votes.get_attribute('innerHTML')
    
    table={'description':full_desc,
           'author_name':author_name,
            'votes_for_winner':votes,
            'Total_votes_for_all':total_votes
            }
    df_books2=df_books2.append(table,ignore_index=True)
df_books2
    
```
